```python
import pandas as pd
import requests
import json
import time
from tqdm import tqdm
```

```python
metadata_api = "https://maayanlab.cloud/sigcom-lincs/metadata-api"
```

```python
libid = "54198d6e-fe17-5ef8-91ac-02b425761653"
```

```python
limit = 10000
skip = 0
signatures = []
while skip < 718055:
    print(skip)
    payload = {
        "limit": limit,
        "skip": skip,
        "fields": ["id", "meta.pert_name", "meta.pert_type", "meta.cmap_id"]
    }
    for i in range(5):
        res = requests.get(metadata_api+"/libraries/%s/signatures"%libid, params={"filter": json.dumps(payload)})
        if res.ok:
            signatures = signatures + res.json()
            break
        else:
            time.sleep(i+1)
    else:
        print(res.text)
    skip += limit
    
```

```python
meta = pd.read_csv("/Users/Andrew/Downloads/siginfo_beta.txt", sep="\t")
```

```python
meta.head()
```

```python
meta = meta.set_index("sig_id")
```

```python
meta.head()
```

```python
meta.columns
```

```python
drug_counts = {}
for i in signatures:
    uid = i["id"]
    cmap_id = i["meta"]["cmap_id"]
    pert_name = i["meta"]["pert_name"]
    brd_id = meta.at[cmap_id, "pert_id"]
    if brd_id not in drug_counts:
        drug_counts[brd_id] = {
            "pert_name": pert_name,
            "sig_count": 0
        }
    drug_counts[brd_id]["sig_count"] += 1
```

```python
drug_meta = pd.read_csv("/Users/Andrew/Downloads/compoundinfo_beta.txt", sep="\t", index_col=0)
```

```python
drug_meta = drug_meta.fillna("-")
```

```python
drug_meta.head()
```

```python
for k, v in tqdm(drug_counts.items()):
    for col in drug_meta.columns:
        if not col == "cmap_name":
            val = drug_meta.at[k, col]
            if not type(val) == str and len(val) > 1:
                drug_counts[k][col] = ", ".join(list(set(val)))
            else:
                drug_counts[k][col] = val
```

```python
drug_tsv = pd.DataFrame.from_dict(drug_counts, orient='index')
drug_tsv.head()
```

```python
drug_tsv = drug_tsv.loc[:,['pert_name', 'target', 'moa', 'canonical_smiles',
       'inchi_key', 'compound_aliases', 'sig_count']]
```

```python
drug_tsv = drug_tsv.sort_values("sig_count", ascending=False)
```

```python
drug_tsv.to_csv("LINCS_small_molecules.tsv", sep="\t")
```
